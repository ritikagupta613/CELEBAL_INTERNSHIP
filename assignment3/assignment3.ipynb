```python
import pandas as pd
```

```python
df= pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Check for missing value
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting the distribution of CGPA
sns.histplot(df['cgpa'], kde=True)
plt.title('CGPA Distribution')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of Placement Exam Marks
sns.histplot(df['placement_exam_marks'], kde=True)
plt.title('Placement Exam Marks Distribution')
plt.xlabel('Placement Exam Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of placements
sns.countplot(x='placed', data=df)
plt.title('Placement Count')
plt.xlabel('Placed')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot of CGPA by Placement Status
sns.boxplot(x='placed', y='cgpa', data=df)
plt.title('CGPA by Placement Status')
plt.xlabel('Placed')
plt.ylabel('CGPA')
plt.show()
```

```python

# Box plot of Placement Exam Marks by Placement Status
sns.boxplot(x='placed', y='placement_exam_marks', data=df)
plt.title('Placement Exam Marks by Placement Status')
plt.xlabel('Placed')
plt.ylabel('Placement Exam Marks')
plt.show()
```
